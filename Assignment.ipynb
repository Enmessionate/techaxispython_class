```python
"""Task 1: The
"Cyber-ID" Generator

Scenario: You are building a registration system
for a futuristic gaming convention.

Instructions:

1. Ask the user for
their username, birth_year (as integer),
and is_member (True/False).

2. Calculate their
current age (assume current year is 2025).

3. Create a
variable entry_fee. If they are a member, fee is 10.5,
otherwise 20.0.

4. Output: Print
a formatted ID card:

`"User: [Name]"""

#Cyber ID generator program
#taking the user input
username=input("Enter your name: ")
birth_year=int(input("Enter your birth year: "))
is_member_input=input("Are you the member of the cyber?(True/False): ").strip().lower()

#converting string values into boolean
is_member = is_member_input == "true"

#for age calculations
current_year = 2025
age = current_year - birth_year

#entry fee determination 
if is_member:
    entry_fee=10.5
else:
    entry_fee=20.0

#print the data for the id card
print("\n-------CYBER ID CARD-------")
print(f"{'Name':<15}: {username}")
print(f"{'Age':<15}: {age}")
print(f"{'Member':<15}: {is_member}")
print(f"{'Entry fee':<15}: Rs.{entry_fee}")
```

```python

"""Task 2: The Smart
Coffee Machine

Scenario: You are programming a vending machine
that checks user credit and sugar preferences.

Instructions:

1. Set a
price: coffee_price = 250.

2. Ask user
for credit_amount and sugar_level (High/Low).

3. Logic:

 - If credit < price: Print
"Insufficient funds."- If credit >= price AND sugar_level is
"High":

Print
"Dispensing Sweet Coffee. Change: [X]"

 - Otherwise:
Print "Dispensing Black Coffee. Change: [X]"
"""

#smart coffee machine program

#set coffee price
coffee_price=250

#take user input
credit = int(input("Enter your credit amount: "))
sugar_level = input("Sugar level (High/Low): ").strip().lower()

if credit<coffee_price:
    print(f"Insufficient funds!")
elif credit>=coffee_price and sugar_level == "high":
    change = credit - coffee_price
    print(f"\nDispensing Sweet Coffee.\nChange: {change}")
else:
    change = credit - coffee_price
    print(f"\nDispensing Black Coffee.\nChange: {change}")
    
```

```python
"""Assignment

Activity 1: Number Guessing Game
Goal: The computer picks a secret number, and the user has to guess it. The program gives hints like "too high" or "too low".
Concepts you can use:
import random (to get a random number)
input() (to get the user's guess)
int() (to convert the user's input string into a number)
while loop (to keep guessing until correct)
if/elif/else (to check the guess and give hints)"""

import random

# Computer picks a secret number between 1 and 100
secret_number = random.randint(1, 100)

# Initialize guess
guess = 0

print("Guess the number between 1 and 100")

# Loop until the guess is correct
while guess != secret_number:
    guess = int(input("Enter your guess: "))

    if guess > secret_number:
        print("Too high!")
    elif guess < secret_number:
        print("Too low!")
    else:
        print("Congratulations! You guessed the correct number.")
```

```python
"""Assignment
Task 1: The messy_data Cleaner
Focus: String Methods, Slicing, F-Strings
Scenario: You have received raw text data from a user input, but it is messy. It has extra spaces, inconsistent capitalization, and contains a forbidden word.
Create a variable raw_text with this exact value:
" PyThOn is AweSome and badword programming! "
Clean the text:
Remove the leading and trailing spaces.
Convert the entire string to lowercase.
Replace the word "badword" with "powerful".
Slice it: Extract the word "programming" from your cleaned text using slicing syntax [start:stop].
Report: Use an f-string to print the final result in this format:
"Original length: {x}, Cleaned text: '{y}', Extracted word: '{z}'"
"""

# Raw text data
raw_text = " PyThOn is AweSome and badword programming! "

# Step 1: Remove leading and trailing spaces
cleaned_text = raw_text.strip()

# Step 2: Convert to lowercase
cleaned_text = cleaned_text.lower()

# Step 3: Replace forbidden word
cleaned_text = cleaned_text.replace("badword", "powerful")

# Step 4: Slice the word "programming"
# "programming" starts at index 27 and ends at 38
extracted_word = cleaned_text[27:38]

# Step 5: Report using f-string
print(
    f"\nOriginal length: {len(raw_text)}, "
    f"\nCleaned text: '{cleaned_text}', "
    f"\nExtracted word: '{extracted_word}'"
)
```
